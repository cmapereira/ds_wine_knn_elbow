```python
X = df.drop(['Class'], axis=1)
```

```python
y = df['Class']
```

```python
X.head()
```

```python
y.head()
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
scaler = StandardScaler()
```

```python
X_train = scaler.fit_transform(X_train)
```

```python
X_test = scaler.transform(X_test)
```

```python
from sklearn.metrics import classification_report,confusion_matrix,accuracy_score,plot_confusion_matrix
```

```python
knn_model = KNeighborsClassifier()
```

```python
test_error_rates = []


for k in range(1,40):
    knn_model = KNeighborsClassifier(n_neighbors=k)
    knn_model.fit(X_train,y_train) 
   
    y_pred = knn_model.predict(X_test)
    
    test_error = 1 - accuracy_score(y_test,y_pred)
    test_error_rates.append(test_error)
```

```python
plt.figure(figsize=(6,4),dpi=100)
plt.plot(range(1,40),test_error_rates,label='Test Error')
plt.legend()
plt.ylabel('Error Rate')
plt.xlabel("K Value")
```

```python
knn_model = KNeighborsClassifier(n_neighbors=32)
```

```python
knn_model.fit(X_train,y_train) 
```

```python
y_pred = knn_model.predict(X_test)
```

```python
confusion_matrix(y_test, y_pred) 
```

```python
plot_confusion_matrix(knn_model, X_test, y_test)
```

```python
print(classification_report(y_test, y_pred))
```

```python
from sklearn.svm import SVC
```

```python
svm = SVC()
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
param_grid = {'C':[0.001,0.01,0.1,0.5,1,2,3], 'kernel':['linear', 'poly', 'rbf'], 'degree':[1,2,3,4]}
```

```python
grid_model_svm = GridSearchCV(svm, param_grid)
```

```python
grid_model_svm.fit(X_train, y_train)
```

```python
grid_model_svm.best_params_
```

```python
svm_pred = grid_model_svm.predict(X_test)
```

```python
confusion_matrix(y_test, svm_pred)
```
